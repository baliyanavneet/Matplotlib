# Subplots

```python
# a diff way to plot graphs

batters.head()
```

```python
# normal way

plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])

plt.title('something')
plt.xlabel('Avg')
plt.ylabel('SR')

plt.show()
```

```python
# another way of plotting graph

fig, ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='D')

ax.set_title('something')
ax.set_xlabel('Avg')
ax.set_ylabel('SR')

fig.show()
```

```python
# plot 2 scatter plots side by side

fid, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize = (10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='green')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike rate')
ax[0].set_ylabel('SR')

ax[1].set_title('Avg Vs Runs')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')
```

```python
# plot 2 scatter plots & 2 histogram plots side by side

fid, ax = plt.subplots(nrows=2,ncols=2,figsize = (10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='green')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)  # (row,col,position)
ax3.hist(batters['avg'])
```

# Contour Plot

- Represent the 3D plot in 2D

```python
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
# plot this ghraph in 2D

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
# filled contour graph

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.sin(yy)

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```
